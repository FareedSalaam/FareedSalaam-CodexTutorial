```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['most_subscribed_youtube_channels (2).csv']))
display(df)
```

```python
df = df.replace(',', '')
display(df)
```

```python
import plotly.express as px
```

```python
##The histogram displays the count of the amount of channels that possess a specific amount of subscibers
```

```python
fig = px.histogram(df, x='Video count', title='Video Count ')
fig.show()
```

```python
# Shows the distribution of categories based on teh amount of channels provided
```

```python
px.pie(df, values='started', names='category', title='Subscribers by Year Started ')
fig.show()
```
